```python
import pandas as pd
from unidecode import unidecode

# Load CSV file
df = pd.read_csv('of.csv')

# Apply unidecode to column names
df.columns = [unidecode(col) for col in df.columns]

# Apply unidecode to column with Unicode text
df['Player'] = df['Player'].apply(unidecode)

# Save updated data to a new CSV file
df.to_csv('outfield_info.csv', index=False, encoding='utf-8')
```
